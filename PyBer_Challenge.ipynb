### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import os
from pandas import pivot

# File to Load (Remember to change these)
city_data_to_load = os.path.join("Resources","city_data.csv")
ride_data_to_load = os.path.join("Resources","ride_data.csv")

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]

suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]

rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]


urban_ride_count = urban_cities_df.groupby(["type"]).count()['ride_id']
suburban_ride_count = suburban_cities_df.groupby(["type"]).count()["ride_id"]
rural_ride_count = rural_cities_df.groupby(["type"]).count()["ride_id"]

# # sum of rides of all city
total_rides_count = pyber_data_df.groupby(['type']).count()['ride_id']
total_rides_count
```

```python
# 2. Get the total drivers for each city type
# urban_drivers_count = urban_cities_df.groupby(["type"]).sum()["driver_count"]
# suburban_drivers_count = suburban_cities_df.groupby(["type"]).count()["driver_count"]
# rural_drivers_count = rural_cities_df.groupby(["type"]).count()["driver_count"]



# #sum total drivers all city
sum_total_drivers_by_type = city_data_df.groupby(["type"]).sum()["driver_count"]
sum_total_drivers_by_type


# Total_drivers = sum(sum_total_drivers_by_type)
# Total_drivers
```

```python
#  3. Get the total amount of fares for each city type
urban_fares_count = urban_cities_df.groupby(["type"]).sum()["fare"]
suburban_fares_count = suburban_cities_df.groupby(["type"]).sum()["fare"]
rural_fares_count = rural_cities_df.groupby(["type"]).sum()["fare"]


#total fares
sum_fares_by_type = pyber_data_df.groupby(["type"]).sum()["fare"]
sum_fares_by_type

# Total_fare = sum(sum_fares_by_type)
# Total_fare
```

```python
#  4. Get the average fare per ride for each city type. 
urban_fare_avg = urban_cities_df.groupby(["fare"]).mean()["ride_id"]
suburban_fare_avg = suburban_cities_df.groupby(["fare"]).mean()["ride_id"]
rural_fare_avg = rural_cities_df.groupby(["fare"]).mean()["ride_id"]

sum_avg_fares_per_ride = pyber_data_df.groupby(["type"]).mean()["fare"]
sum_avg_fares_per_ride

#avg fare per ride by city type
# sum of all fares divided by total rides

avg_fare_per_city = sum_fares_by_type/total_rides_count
avg_fare_per_city
```

```python
# 5. Get the average fare per driver for each city type. 
urban_fare_driver_avg = urban_cities_df.groupby(["type"]).mean()["driver_count"]
suburban_fare_driver_avg = suburban_cities_df.groupby(["type"]).mean()["driver_count"]
rural_fare_driver_avg = rural_cities_df.groupby(["type"]).mean()["driver_count"]

#avg_fare_per_driver = city_data_df.groupby(["type"]).mean()["driver_count"]
# avg fare per driver
# sum of all fares divided by total drivers


avg_fare_per_driver = sum_fares_by_type/sum_total_drivers_by_type
avg_fare_per_driver

```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({'Total Rides':total_rides_count, 'Total Drivers':sum_total_drivers_by_type, 'Total Fares':sum_fares_by_type, 'Average Fare per Ride':avg_fare_per_city, 'Average Fare per Driver':avg_fare_per_driver,})
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,}".format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,}".format)
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:.2f}".format)

pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
sum_fares_date_df = pyber_data_df.groupby(['type', 'date']).sum()['fare'].to_frame()
sum_fares_date_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
sum_fares_date_df.reset_index()
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pivot_table = sum_fares_date_df.pivot_table(index=['date'], columns ='type', values='fare')
pivot_table
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
pivot_df = pivot_table.loc['2019-01-01':'2019-04-29']
pivot_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
pivot_table.index = pd.to_datetime(pivot_table.index)
pivot_table
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
pivot_table.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
pivot_table = pivot_table.resample('W').sum()
pivot_table.head()
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

```

```python
pivot_table.plot(figsize=(20,8),)

plt.xlabel('')
plt.ylabel('Fare ($USD)')
plt.title('Total Fare by City Type')
plt.savefig('Resources/Pyber_fare_summary.png', dpi = 100)
plt.show()
```
